```python
#  colors in scatter plot
```

```python
x = np.random.rand(50)
y = np.random.rand(50) 
colors = np.random.rand(50)

plt.scatter(x,y, c = colors)


plt.xlabel("This is my x-Axis")
plt.ylabel("This is my y-Axis")
plt.title("This is my Title")


plt.show()
```

```python
#colors
```

```python
plt.scatter(x,y, c = colors)
```

```python
#size
```

```python
sizes = 1000 * np.random.rand(50)
plt.scatter(x,y, c = colors, s = sizes)
```

```python
#alpha
```

```python
plt.scatter(x,y, c = colors, s = sizes, alpha=.3)
```

```python
#bar graph (horizontal)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = ['a', 'b', 'c', 'd', 'e']
y = np.random.rand(5)

plt.barh(x,y)     #only instead of bar write 'barh'


plt.title("This is the Title")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
#figsize
```

```python
plt.figure(figsize=(6,2))
plt.xlabel("This is my x-Axis")
plt.ylabel("This is my y-Axis")
plt.title("This is my Title")
plt.plot(x,y, '--b')
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50) 
colors = np.random.rand(50)


sizes = 1000 * np.random.rand(50)
plt.scatter(x,y, c = colors, s = sizes, alpha=.4, marker = 'x')

plt.xlabel("This is my x-Axis")
plt.ylabel("This is my y-Axis")
plt.title("This is my Title")


plt.show()
```
