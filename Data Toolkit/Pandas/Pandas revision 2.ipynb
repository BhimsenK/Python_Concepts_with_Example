```python
import pandas as pd

# Sample DataFrame
data = {
    'EmployeeID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank', 'Grace', 'Henry'],
    'Department': ['HR', 'Finance', 'HR', 'IT', 'Finance', 'IT', 'HR', 'IT'],
    'Salary': [50000, 60000, 55000, 70000, 65000, 75000, 60000, 80000]
}

df3 = pd.DataFrame(data)
print(df3)
```

```python
import pandas as pd

# Sample DataFrame 1
data1 = {
    'ID': [101, 102, 103],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
}
df1 = pd.DataFrame(data1)

# Sample DataFrame 2
data2 = {
    'ID': [102, 103, 104],
    'Department': ['HR', 'Finance', 'IT'],
    'Salary': [50000, 60000, 70000]
}
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
#Left Join  ()

'''
"A left join includes all elements from the left set and only matching elements from the right set.
'''
```

```python
left_join = pd.merge(df1, df2, on = 'ID', how = 'left')
```

```python
left_join
```

```python
#Right Join  ()

'''
A right join includes all elements from the right set and only matching elements from the left set."
'''
```

```python
right_join = pd.merge(df1, df2, on = 'ID', how = 'right')
```

```python
right_join
```

```python
#Inner Join  ()

'''
An inner join includes only the common elements between two sets
'''
```

```python
# Inner Join
inner_join = pd.merge(df1, df2, on='ID', how='inner')
```

```python
inner_join
```

```python
#outer Join  ()

'''
An outer join includes all elements from both sets, filling in missing values with NaN where there are no matches
'''
```

```python
outer_join = pd.merge(df1, df2, on='ID', how='outer')
```

```python
outer_join
```

```python
#------------ Joins ------------- 
```

```python
import pandas as pd

# Sample DataFrame 1
data1 = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
}
df1 = pd.DataFrame(data1)

# Sample DataFrame 2
data2 = {
    'Name': ['Bob', 'Charlie', 'David'],
    'Department': ['HR', 'Finance', 'IT'],
    'Salary': [50000, 60000, 70000]
}
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python

# Left Join
left_join_df = df1.join(df2.set_index('Name'), on='Name', how='left')
print("Left Join:")
print(left_join_df)
print()

# Right Join
right_join_df = df1.join(df2.set_index('Name'), on='Name', how='right')
print("Right Join:")
print(right_join_df)
print()

# Inner Join
inner_join_df = df1.join(df2.set_index('Name'), on='Name', how='inner')
print("Inner Join:")
print(inner_join_df)
print()

# Outer Join
outer_join_df = df1.join(df2.set_index('Name'), on='Name', how='outer')
print("Outer Join:")
print(outer_join_df)
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'Student': ['Alice', 'Bob', 'Charlie'],
    'Marks': [45, 55, 85]
}
df = pd.DataFrame(data)

# Define a function to map marks to strings
def map_marks(marks):
    if marks < 50:
        return 'Try hard'
    elif marks <= 80:
        return 'Ok'
    else:
        return 'Keep on going'
```

```python
df
```

```python
# Apply the function to create a new column with strings
df['Remarks'] = df['Marks'].apply(map_marks)
```

```python
#import pandas as pd

# Sample DataFrame
data = {
    'Value': [10, 20, 30, 40, 50, 60, 70, 80]
}
df = pd.DataFrame(data)

# Create a rolling window object with window size 3
rolling_mean = df['Value'].rolling(window=3)

# Compute the rolling mean
df['Rolling_Mean'] = rolling_mean.mean()

print("DataFrame with Rolling Mean:")
print(df)
```
