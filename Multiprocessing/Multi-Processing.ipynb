```python
import time
start = time.perf_counter()

def say_my_name():
    print(f'I am Hysenberg !!!')
    time.sleep(2)
    
say_my_name()
say_my_name()

end = time.perf_counter()

print(f'Time taken = {round(end-start)} Sec.')
```

```python
#with Multi-Processing
```

```python
import time, multiprocessing
start = time.perf_counter()

def say_my_name():
    print(f'I am Hysenberg !!!')
    time.sleep(2)

p1 = multiprocessing.Process(target = say_my_name)
p2 = multiprocessing.Process(target = say_my_name)
# say_my_name()
# say_my_name()

p1.start()
p2.start()

#Mentioning these join() at end cause if we mention this directly after 1st P1.start() then it will wait till p1 to end. 
p1.join()
p2.join()

end = time.perf_counter()

print(f'Time taken = {round(end-start)} Sec.')
```

```python
#with Multi-Processing using loop
```

```python
#for 1000 > 15 sec.
```

```python
import time, multiprocessing
start = time.perf_counter()

def say_my_name():
    print(f'I am Hysenberg !!!')
    time.sleep(2)


Processes = []
for i in range(1000):
    p = multiprocessing.Process(target = say_my_name)
    p.start()
    Processes.append(p)

for proc in Processes:
    proc.join()
    
end = time.perf_counter()

print(f'Time taken = {round(end-start)} Sec.')
```
